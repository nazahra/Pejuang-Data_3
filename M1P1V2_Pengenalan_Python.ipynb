# Pengenalan Python

Python adalah bahasa pemrograman tingkat tinggi dengan dynamic semantics, bersifat objek oriented, dan interpretatif.Bahasa pemrograman ini fokus dikembangkan pada keterbacaan bahasa pemrograman.
1. **Bahasa Pemrograman Tingkat Tinggi (High Level Programming)** artinya sintaks lebih mudah ditafsirkan oleh manusia.
2. **Interpretatif** artinya bahasa ini memerlukan intepreter atau penerjemah. Ini memungkinkan bagi kita untuk menggunakan bahasa yang mudah kita pahami.
3. **Object Oriented** ini berarti bahwa bahasa pemrograman ini berorientasi pada objek. Hal ini membuat Phyton lebih muda untuk dikembangkan.
4. **Dynamic Semantic** Phyton secara otomatis akan menyesuaikan tipe dari suatu variabel berdasarkan input.

## Mengapa Phyton

1. Mudah dipahami.
2. Gratis & open source.
3. Dinamis.
4. Memiliki library terlengkap untuk data sains.

## Struktur Program pada Python

Phyton pada umumnya terdiri dari:
1. **Statement**

```python
print("belajar Python menyenangkan")
print("halo dunia")
```

2. **Variables**

```python
bilangan1 = 5
bilangan2 = 10
kalimat = "Pejuang Data 3"
```

3. **Literals** simbol-simbol yang dapat kita gunakanuntuk mengisi sebuah variabel.

4. **Operators** simbol-simbol yang dapat digunakan untuk mengubah nilai dari suatu variabel dengan melibatkan satu atau lebih variabel dan literal.

```python
bilangan1 + bilangan2
```

5. **Reserved Words** kumpulan kata-kata yang memiliki makna khusus dalan bahasa pemrograman Python. Kata `False`, `Return`, dan `For` merupakan contoh Reserved Words

6. **Whitespace** pada bahasa Python, spasi dan tab memiliki makna khusus untuk menandai serangkaian blok dalam kode Python.

7. **Comments** merupakan sekumpulan teks yang dituliskan di dalam sebuah program yang tidak akan memengaruhi hasil dari sebuah program.
Terdapat 2 jenis comments di dalam Phyton, yaitu:

**a. Single Line Comments** (komen dalam satu baris) seperti pada contoh berikut

```python
# Ini adalah comment
print("belajar Python")
```

**b. Multi Line Comments** (komen dalam beberapa baris) yang dapat dituliskan seperti pada contoh berikut:

```python
'''
Ini adalah komentar panjang
Bisa digunakan untuk memberikan penjelasan
panjang pada Sell Kode Python,
Misal dokumentasi program, spesifikasi
input output atau lainnya
'''

print('belajar Python menyenangkan')
```
